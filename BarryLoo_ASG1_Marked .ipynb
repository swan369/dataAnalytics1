Write a user-defined function that has three parameters: a year value, an numpy array input, and a column index. It will sum the values in that column index for the year specified and return the result.

Use this user-defined function to answer the following question:

Consider the following 9 year values:
	2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011
 
Print the top 3 years with the highest trade values.

```python
# Answer to Q3 (15 marks)
import math
import numpy as np
myFile = np.genfromtxt("trade_international.csv", delimiter = ",", skip_header=1)

def calcTotal (year, npArray, colIndex):
    """return sum of values in colIndex of year specified"""
    try:
        npResults = np.array(list(filter(lambda x: x[0] == float(year) and not np.isnan(x[1]) , npArray)))
        final = np.sum(npResults, axis = 0)[colIndex]
        return math.floor(final)

    except:
        return 0

yearList= [2003,2004,2005,2006,2007,2008,2009,2010,2011]  # could use range eg. yearList = range(2003, 2012)  P.Loy
yearValues = []
for year in yearList:
    total = calcTotal(year,myFile,1)
    yearValues.append([year,total])

npYearValues = np.array(yearValues)

npHighSort = npYearValues[np.argsort(npYearValues[:,1])[::-1]]
# could use a loop to print.  P.Loy
print(f"""The top 3 most popular years: 
In {npHighSort[0,0]}, there were ${npHighSort[0,1]:,} of trade transactions.
In {npHighSort[1,0]}, there were ${npHighSort[1,1]:,} of trade transactions.
In {npHighSort[2,0]}, there were ${npHighSort[2,1]:,} of trade transactions.
""")
```
